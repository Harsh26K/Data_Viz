```python
import numpy as np 
import pandas as pd
```

```python
dataset1 = pd.read_csv('googleplaystore.csv')
dataset2 = pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
dataset1.columns
```

```python
AppSet = set(dataset2.App.values)
AppSet
```

```python
Applist = list(AppSet)
len(AppSet)
```

```python
dataset1.shape[1]
```

```python
Category = []
App=[]
i=0
for i in range(0,dataset1.shape[0]):
    value = dataset1['App'][i]
    if value in Applist and ( value not in App):
        App.append(dataset1['App'][i])
        Category.append(dataset1['Category'][i])
```

```python
App
```

```python
Category
```

```python
d={'App':App,'Category':Category}
new_data = pd.DataFrame(data=d)
```

```python
new_data.to_csv('new_data.csv')
```
